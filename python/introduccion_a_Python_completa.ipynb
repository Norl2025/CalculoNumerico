# Graficar funciones

## Objetivo: graficar sen(x) entre -5 y 7

```python
import numpy as np #paquete para manejar arreglos
from matplotlib import pyplot as plt #para graficar

# Detalle técnico: no darle bola.
plt.rcParams["figure.figsize"] = (10,7) #tamaño de los gráficos: ancho x alto
```

```python
# Aquí se puede discutir 5 min. la ventaja de definir funciones.

#definir la funcion sen(x)
def func(x):
  y = np.sin(x)
  return y

# algunos ejemplos de uso llamándola
resultado = func(2)
print(resultado)

resultado2 = func(np.pi/2)
print(resultado2)

xs = np.array([np.pi, np.pi/2, 2])
ys = func(xs)
print(xs)
print(ys)
```

```python
# array de valors de x (con linspace por ej.)
valores_x = np.linspace(-5,7,1000)
print(valores_x)
```

```python
# ídem para valores de y
# observar que se evalúa la función en todo un vector de valores

valores_y = func(valores_x)
print(valores_y)
```

```python
# gráfico de la función bien básico
plt.plot(valores_x, valores_y)
```

```python
plt.plot(valores_x, valores_y, '*')
```

```python
plt.plot(valores_x, valores_y, '*')
plt.xlim(0, 1) #zoom en eje x para que se noten los puntos
```
